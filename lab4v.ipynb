```python
import sys
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

plt.rcParams["figure.figsize"] = [6, 4]
image = cv.imread('my.jpg')
# Использование библиотеки matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#%matplotlib inline позволяет выводить графики matplotlib в Jupyter
plt.imshow(image)
```

```python
image1 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image1)
```

```python
import cv2
thresholds = [90, 160]
prev_image = image.copy()

for tr in thresholds:
    _, prev_image = cv2.threshold(prev_image, tr, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(14, 10))
plt.imshow(prev_image, cmap='gray')
plt.show()
```

```python
cv.imwrite('bi.jpg', prev_image)
```
